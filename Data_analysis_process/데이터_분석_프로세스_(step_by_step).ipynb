````python
# Target label ÏÉùÏÑ± Sample Code(1-loop)
i = '2011-01'

# Íµ¨ÌïòÍ≥†Ïûê ÌïòÎäî Í∏∞Ï§ÄÏõî Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
df_left = df[df['bsym'] == i]

# Í∏∞Ï§ÄÎÖÑÏõî + 1m
bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')

# Í∏∞Ï§ÄÎÖÑÏõî + 1m, CustomerIDÎäî 1ÏõîÏùò Íµ¨Îß§Ìïú Í≥†Í∞ù
df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique()) # bsym_1mÏóê Ìï¥ÎãπÌïòÎäî ÏõîÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Í≥†Ïú†Ìïú CustomerIDÎ•º Ï∂îÏ∂úÌï©ÎãàÎã§.
df_right['target'] = 'Y'
df_right.columns = ['CustomerID', 'target']

# Í∏∞Ï§ÄÎÖÑÏõî DataÏóê Í∏∞Ï§ÄÎÖÑÏõî +1M Í≥†Í∞ùÏùÑ JoinÌïòÎ©¥ Target Data ÏÉùÏÑ± ÏôÑÎ£å
df_merge = pd.merge(df_left, df_right, how='left')
df_merge['target'] = df_merge['target'].fillna('N')
````

````python
# forÎ¨∏ ÌôúÏö© Ï†ÑÏ≤¥ Data Set ÎåÄÏÉÅ Target label ÏÉùÏÑ±
loop_list = list(df['bsym'].unique())
df_all = []

for i in loop_list :
  df_left = df[df['bsym'] == i]
  bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')
  df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique())
  df_right['target'] = 1
  df_right.columns = ['CustomerID', 'target']
  df_merge = pd.merge(df_left, df_right, how='left')
  df_merge['target'] = df_merge['target'].fillna(0)

  df_all.append(df_merge)

df_all = pd.concat(df_all)
````

### **ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞: ÏóÖÏÉòÌîåÎßÅÍ≥º Îã§Ïö¥ÏÉòÌîåÎßÅ**
- **ÏóÖÏÉòÌîåÎßÅ(Up-sampling)**: Ï£ºÏñ¥ÏßÑ ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Î•º Î≥¥Í∞ÑÌïòÏó¨ Îçî ÎÜíÏùÄ ÎπàÎèÑÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.
- **Îã§Ïö¥ÏÉòÌîåÎßÅ(Down-sampling)**: Ï£ºÏñ¥ÏßÑ ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Î•º ÏßëÍ≥ÑÌïòÏó¨ Îçî ÎÇÆÏùÄ ÎπàÎèÑÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§.

### **ÎπÑÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞: SMOTEÏôÄ Ï∏µÌôî ÏÉòÌîåÎßÅ**
- **SMOTE (Synthetic Minority Over-sampling Technique)**: ÏÜåÏàò ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Î•º Ïò§Î≤ÑÏÉòÌîåÎßÅÌïòÏó¨ Í∑†ÌòïÏûàÍ≤å ÎßåÎì≠ÎãàÎã§.
- **Ï∏µÌôî ÏÉòÌîåÎßÅ(Stratified Sampling)**: Í∞Å Í≥ÑÏ∏µÏóêÏÑú ÏÉòÌîåÏùÑ ÎèôÏùºÌïú ÎπÑÏú®Î°ú Ï∂îÏ∂úÌïòÏó¨ Ï†ÑÏ≤¥ ÏÉòÌîåÏùÑ ÏñªÏäµÎãàÎã§.

#### 2. **ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥ÏÉòÌîåÎßÅ**

#### 3. **SMOTE Ïò§Î≤ÑÏÉòÌîåÎßÅ**
`SMOTE`Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÏÜåÏàò ÌÅ¥ÎûòÏä§Î•º Ïò§Î≤ÑÏÉòÌîåÎßÅÌï©ÎãàÎã§. Ïù¥Î•º ÌÜµÌï¥ `target` ÌÅ¥ÎûòÏä§Ïùò Î∂ÑÌè¨Í∞Ä 1:1Ïù¥ Îê©ÎãàÎã§.

#### 4. **Ï∏µÌôî ÏÉòÌîåÎßÅ**

> **Feature EngineeringÏù¥ÎûÄ?**

```
* Î™®Îç∏ÎßÅÏùò Î™©Ï†ÅÏù∏ ÏÑ±Îä•Ìñ•ÏÉÅÏùÑ ÏúÑÌï¥ FeatureÎ•º ÏÉùÏÑ±, ÏÑ†ÌÉù, Í∞ÄÍ≥µÌïòÎäî ÏùºÎ†®Ïùò Î™®Îì† ÌôúÎèô
```


> **Target Î≥ÄÏàòÏôÄÏùò ÏùòÎØ∏ÏûàÎäî Î≥ÄÏàò ÏÑ†ÌÉùÌïòÎäî Î∞©Î≤ï**

```
* Regression(ÌöåÍ∑Ä) : ÏÉÅÍ¥ÄÍ≥ÑÏàò Î∂ÑÏÑùÏùÑ ÌÜµÌï¥ Ïú†ÏùòÎØ∏Ìïú Î≥ÄÏàò ÏÑ†ÌÉù

```

> **IV (Information Value)**

```
* Feature ÌïòÎÇòÍ∞Ä Good(Target)Í≥º Bad(Non-target)ÏùÑ Ïûò Íµ¨Î∂ÑÌï¥Ï§Ñ Ïàò ÏûàÎäîÏßÄÏóê ÎåÄÌïú Ï†ïÎ≥¥ÎüâÏùÑ ÌëúÌòÑÌïòÎäî Î∞©Î≤ïÎ°†
* IV ÏàòÏπòÍ∞Ä ÌÅ¥ÏàòÎ°ù TargetÍ≥º Non-targetÏùÑ Ïûò Íµ¨Î∂ÑÌï† Ïàò ÏûàÎäî Ï†ïÎ≥¥ÎüâÏù¥ ÎßéÏùÄ FeatureÏù¥Í≥†, IV ÏàòÏπòÍ∞Ä ÏûëÏùÑÏàòÎ°ù Ï†ïÎ≥¥ÎüâÏù¥ Ï†ÅÏùÄ Feature
* (Target data Íµ¨ÏÑ±ÎπÑ - Non-target data Íµ¨ÏÑ±ÎπÑ) * WoE
* WoE(Weight of Evidence) = ln(Target data Íµ¨ÏÑ±ÎπÑ / Non-target data Íµ¨ÏÑ±ÎπÑ)
* ln(ÏûêÏó∞Î°úÍ∑∏)Î•º Ï∑®ÌïòÎäî Ïù¥Ïú†? : WoEÏùò ÏµúÎåÄ, ÏµúÏÜåÍ∞íÏùò Î≤îÏúÑÎ•º ÎßûÏ∂∞Ï£ºÍ∏∞ ÏúÑÌï¥ (1~‚àû , -‚àû~1)
```
