````python
# Target label ÏÉùÏÑ± Sample Code(1-loop)
i = '2011-01'

# Íµ¨ÌïòÍ≥†Ïûê ÌïòÎäî Í∏∞Ï§ÄÏõî Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
df_left = df[df['bsym'] == i]

# Í∏∞Ï§ÄÎÖÑÏõî + 1m
bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')

# Í∏∞Ï§ÄÎÖÑÏõî + 1m, CustomerIDÎäî 1ÏõîÏùò Íµ¨Îß§Ìïú Í≥†Í∞ù
df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique()) # bsym_1mÏóê Ìï¥ÎãπÌïòÎäî ÏõîÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Í≥†Ïú†Ìïú CustomerIDÎ•º Ï∂îÏ∂úÌï©ÎãàÎã§.
df_right['target'] = 'Y'
df_right.columns = ['CustomerID', 'target']

# Í∏∞Ï§ÄÎÖÑÏõî DataÏóê Í∏∞Ï§ÄÎÖÑÏõî +1M Í≥†Í∞ùÏùÑ JoinÌïòÎ©¥ Target Data ÏÉùÏÑ± ÏôÑÎ£å
df_merge = pd.merge(df_left, df_right, how='left')
df_merge['target'] = df_merge['target'].fillna('N')
````

````python
# Target data ÏÉùÏÑ± ÌôïÏù∏
df_merge.head()
````

````python
# forÎ¨∏ ÌôúÏö© Ï†ÑÏ≤¥ Data Set ÎåÄÏÉÅ Target label ÏÉùÏÑ±
loop_list = list(df['bsym'].unique())
df_all = []

for i in loop_list :
  df_left = df[df['bsym'] == i]
  bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')
  df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique())
  df_right['target'] = 1
  df_right.columns = ['CustomerID', 'target']
  df_merge = pd.merge(df_left, df_right, how='left')
  df_merge['target'] = df_merge['target'].fillna(0)

  df_all.append(df_merge)

df_all = pd.concat(df_all)
````
