## **01. 문제해결 프로세스 단계 소개**
---
> **문제해결 프로세스란?**

```
현업의 이슈를 데이터 분석으로 해결하기 위한 프로세스
※ 현업의 이슈 : 데이터 기반 신규 프로젝트 기획, 핵심 이슈 해결 등
```

> **Step 1. Define The Problem (문제정의)**

```
* 현재 기업의 문제상황을 문제현상과 문제발생으로 인한 기업의 피해로 나누어 정의함
* 과제의 개요를 작성하는 단계로서 문제현상을 데이터로 설명할 수 있도록 함
```
> **Step 2. Expected Effects (기대효과)**

```
* 'Step1'에서 정의한 문제현상을 해결했을 때 기대할 수 있는 효과를 정의함
* 기대효과는 정성적이 아닌 정량적으로 작성할 수 있어야 함
ex) 신규 고객의 증가 (X) → 신규 고객 증가로 얻는 수익, + xx억 (O)
```

> **Step 3. Solution (해결방안)**

```
* 우리가 정의한 문제를 해결했을 때 얻을 수 있는 기대효과가 미미하다면 해당 과제의 Drop을 고려할 수 도 있음
* 기대효과가 크고, 진행해야한다고 판단되는 과제라고 한다면, 해결하기 위한 방안을 탐색하고 List up함
* 해결방안은 꼭 모델링이 아니여도 됨 (※ 간단한 통계분석으로도 해결할 수 있음)
```

> **Step 4. Prioritize (우선순위)**

```
* List-up한 해결방안들에 대해서 우선순위를 설정함
* 빠르게 수행할 수 있고, 결과를 파악할 수 있는 방법을 가장 최우선적으로 실행함
* 1순위 방법론으로 해결이 어려울시를 대비하여 2, 3순위 방법론도 준비
```

> **Step 5.  Data Analysis (분석)**

```
* 'Step4'에서 선정한 Solution을 활용하여 데이터 분석을 진행
```

> **Step 6. Performance Monitoring (성과측정)**

```
* 'Step1'에서 정의한 문제의 상황이 우리의 Solution을 통해서 좋아졌는지 측정하는 단계
* As-is vs To-be로 비교를 진행하며, 정량적인 지표로 비교할 수 있도록 준비

```

> **Step 7. Model Operation (모델운영)**

```
* 최종적으로 완성된 Solution을 현업의 상황에 맞춰 주기적으로 운영하기 위한 프로세스를 설계함
* 일, 주, 월 주기를 설정해야하고, 주기적으로 모델 업데이트 일정도 고려해야 함
* 모델 운영 중 오류가 발생했을 시 처리하기 위한 프로세스 수립도 필요함
```

````python
from datetime import datetime

# 기준년월 추출 (bsym)
df['date'] = df['InvoiceDate'].str.split(' ').str[0]
df['date'] = pd.to_datetime(df['date'])
df['bsym'] = df['date'].apply(lambda x : datetime.strftime(x, '%Y-%m'))

# CustomerID는 str Type으로 변경
df['CustomerId'] = df['CustomerID'].astype(str) # str로 변경시킬까? ID라서?
df.head()
````

````python
# 기준년월 추가한 원본 데이터 copy
df_origin = df.copy()
````

````python
# 데이터 적재 기간 확인
df['date'].min(), df['date'].max()
````

````python
# 기준년월 기준으로 Unique한 CustomerID 추출
# 매 월 구매한 Unique한 고객을 파악하기 위함
df = df[['bsym', 'CustomerID']].drop_duplicates(keep='first')
df.head()
````

````python
# Target label 생성 Sample Code(1-loop)
i = '2011-01'

# 구하고자 하는 기준월 데이터 추출
df_left = df[df['bsym'] == i]

# 기준년월 + 1m
bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')

# 기준년월 + 1m, CustomerID는 1월의 구매한 고객
df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique()) # bsym_1m에 해당하는 월의 데이터에서 고유한 CustomerID를 추출합니다.
df_right['target'] = 'Y'
df_right.columns = ['CustomerID', 'target']

# 기준년월 Data에 기준년월 +1M 고객을 Join하면 Target Data 생성 완료
df_merge = pd.merge(df_left, df_right, how='left')
df_merge['target'] = df_merge['target'].fillna('N')
````

````python
# Target data 생성 확인
df_merge.head()
````

````python
df_merge['target'].value_counts()
````

````python
# for문 활용 전체 Data Set 대상 Target label 생성
loop_list = list(df['bsym'].unique())
df_all = []

for i in loop_list :
  df_left = df[df['bsym'] == i]
  bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')
  df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique())
  df_right['target'] = 1
  df_right.columns = ['CustomerID', 'target']
  df_merge = pd.merge(df_left, df_right, how='left')
  df_merge['target'] = df_merge['target'].fillna(0)

  df_all.append(df_merge)

df_all = pd.concat(df_all)
````

이 코드는 `df` DataFrame에서 각 고유한 월(`bsym`)에 대해서, 각 월의 고객 중 다음 달에도 구매한 고객들에게는 타겟 라벨로 1을, 다음 달에 구매하지 않은 고객들에게는 0을 부여합니다. 이 과정을 모든 고유한 월에 대해 수행한 후, 모든 결과를 하나의 DataFrame으로 합칩니다.

### 코드 설명
1. **초기 설정:**
```python
loop_list = list(df['bsym'].unique())
df_all = []
```
   - `loop_list`에 `df`의 `'bsym'` 열의 고유한 값들을 리스트 형태로 저장합니다. 이 리스트에는 고유한 년월이 저장됩니다.
   - `df_all`은 각 루프에서 생성된 DataFrame들을 저장할 리스트입니다.

2. **For 루프:**
```python
for i in loop_list:
```
   - `loop_list`의 각 원소, 즉 각 고유한 년월에 대하여 루프를 수행합니다.

3. **루프 내 연산:**
```python
df_left = df[df['bsym'] == i]
bsym_1m = (pd.to_datetime(i) + pd.DateOffset(months=1)).strftime('%Y-%m')
df_right = pd.DataFrame(df[df['bsym'] == bsym_1m]['CustomerID'].unique())
df_right['target'] = 1
df_right.columns = ['CustomerID', 'target']
df_merge = pd.merge(df_left, df_right, how='left')
df_merge['target'] = df_merge['target'].fillna(0)
```
   - `df_left`는 현재 월(`i`)의 데이터를 저장합니다.
   - `bsym_1m`은 현재 월(`i`)에서 한 달 후의 년월을 계산하여 저장합니다.
   - `df_right`는 `bsym_1m`에 해당하는 월의 고유한 `CustomerID`와 타겟 라벨 `1`을 포함하는 DataFrame을 생성합니다.
   - `df_merge`는 `df_left`와 `df_right`를 왼쪽 조인합니다. 이렇게 하면 현재 월에 구매한 모든 고객에 대해 다음 달에 구매한 고객은 `target`이 1, 그렇지 않은 고객은 NaN값이 들어갑니다.
   - `fillna(0)`을 사용하여 `target` 열의 NaN값을 0으로 채웁니다.

4. **결과 저장:**
```python
df_all.append(df_merge)
```
   - `df_all` 리스트에 각 루프에서 얻은 `df_merge` DataFrame을 추가합니다.

5. **결과 합치기:**
```python
df_all = pd.concat(df_all)
```
   - `pd.concat(df_all)`을 사용하여 `df_all` 리스트에 저장된 모든 DataFrame들을 하나의 DataFrame으로 합칩니다.

### 최종 결과

## **03. Data Sampling (2)**
---

이 코드 부분은 데이터 샘플링에 중점을 둔 것으로 보입니다. 여기서는 시계열 데이터의 업샘플링과 다운샘플링, 그리고 비시계열 데이터의 SMOTE 오버샘플링과 층화 샘플링이 이루어집니다.

### **시계열 데이터: 업샘플링과 다운샘플링**
- **업샘플링(Up-sampling)**: 주어진 시계열 데이터를 보간하여 더 높은 빈도로 변환합니다.
- **다운샘플링(Down-sampling)**: 주어진 시계열 데이터를 집계하여 더 낮은 빈도로 변환합니다.

### **비시계열 데이터: SMOTE와 층화 샘플링**
- **SMOTE (Synthetic Minority Over-sampling Technique)**: 소수 클래스의 인스턴스를 오버샘플링하여 균형있게 만듭니다.
- **층화 샘플링(Stratified Sampling)**: 각 계층에서 샘플을 동일한 비율로 추출하여 전체 샘플을 얻습니다.

### 코드 분석 및 리뷰
#### 1. **시계열 데이터 업샘플링**
코드는 주어진 시간 인덱스를 기반으로 데이터프레임 `df_time`을 생성하고, 이를 업샘플링합니다. 다양한 방법으로 NaN 값을 채울 수 있습니다. (`ffill`, `bfill`, `0`, `mean`, `interpolate`)

#### 2. **시계열 데이터 다운샘플링**
업샘플링된 데이터프레임 `df_upsample`을 다시 원래의 빈도로 다운샘플링합니다.

#### 3. **SMOTE 오버샘플링**
`SMOTE`를 사용하여 소수 클래스를 오버샘플링합니다. 이를 통해 `target` 클래스의 분포가 1:1이 됩니다.

#### 4. **층화 샘플링**
층화 샘플링은 각 월별로 수행됩니다. 각 월별로 샘플링된 데이터는 `df_all_sample`에 병합됩니다. 층화 샘플링이 정상적으로 수행되었는지 확인하기 위해 샘플링된 데이터의 `target_ratio`와 원래 데이터의 `target_ratio`를 비교합니다.

### 코드 개선 사항
1. **층화 샘플링 코드 개선**: 각 월별로 개별적으로 샘플링하는 대신, `groupby`와 `apply`를 활용하여 더 간결하게 코딩할 수 있습니다.

해당 코드는 Python에서 pandas 라이브러리를 사용하여 데이터 프레임을 처리하고 있습니다. 아래는 코드를 단계별로 분석한 내용입니다.

### 1. **Groupby**
```python
df_mart = df_origin_sample.groupby(['bsym', 'CustomerID'])
```
- `df_origin_sample` 데이터 프레임을 `bsym`과 `CustomerID`로 그룹화합니다.
- 이것은 같은 'bsym'과 'CustomerID' 값을 가진 여러 행들을 하나의 그룹으로 묶을 것입니다.

### 2. **Aggregation**
```python
['amt'].agg(total_amt = ('sum'))
```
- 각 그룹에 대해, 'amt' 열의 값을 합산(`sum`)하여 새로운 'total_amt' 열을 생성합니다.
- 예를 들어, 같은 'bsym'과 'CustomerID'를 가진 행이 여러 개 있을 때, 그 행들의 'amt' 값들을 모두 합한 값을 'total_amt'에 저장합니다.

### 3. **Reset Index**
```python
.reset_index()
```
- `groupby`와 `agg`의 결과로 나온 데이터 프레임의 인덱스를 리셋합니다. 이는 기존에 인덱스로 설정되어 있던 'bsym'과 'CustomerID'를 일반 열로 바꾸고, 새로운 디폴트 인덱스(0, 1, 2, ...)를 생성합니다.

### 4. **결과**
```python
df_mart
```
- 최종적으로, `df_mart` 데이터 프레임에는 각 'bsym'과 'CustomerID' 그룹에 대한 'total_amt' 값이 포함되어 있을 것입니다.

### 주의
해당 코드에는 작은 오류가 있을 수 있습니다. `agg` 함수에서 딕셔너리를 사용할 때, 일반적으로 다음과 같이 작성합니다.

```python
agg({'column_name': 'aggregation_function'})
```


## **03. Integer Feature Engineering**

> **Feature Engineering이란?**

```
* 모델링의 목적인 성능향상을 위해 Feature를 생성, 선택, 가공하는 일련의 모든 활동
```


> **Target 변수와의 의미있는 변수 선택하는 방법**

```
* Regression(회귀) : 상관계수 분석을 통해 유의미한 변수 선택
* Classification(분류) : bin(통)으로 구분 후 Target 변수와의 관계 파악
```

> **IV (Information Value)**

```
* Feature 하나가 Good(Target)과 Bad(Non-target)을 잘 구분해줄 수 있는지에 대한 정보량을 표현하는 방법론
* IV 수치가 클수록 Target과 Non-target을 잘 구분할 수 있는 정보량이 많은 Feature이고, IV 수치가 작을수록 정보량이 적은 Feature
* (Target data 구성비 - Non-target data 구성비) * WoE
* WoE(Weight of Evidence) = ln(Target data 구성비 / Non-target data 구성비)
* ln(자연로그)를 취하는 이유? : WoE의 최대, 최소값의 범위를 맞춰주기 위해 (1~∞ , -∞~1)
```

````python
# for문 활용 자동화
integer_list = ['total_amt', 'max_amt', 'min_amt', 'total_cnt','max_cnt', 'min_cnt', 'total_qty', 'max_qty', 'min_qty']
iv_df = []

for val in integer_list :
    # ▶ Binning
    quantile  = df_mart[val].quantile(q=list(np.arange(0.05,1,0.05)), interpolation = 'nearest')

    # ▶ 그룹 초기화
    df_mart['grp'] = 0

    # ▶ 40%, 75% 기준으로 총 3개 구간화 진행
    for i in  range(len(df_mart[val])) :
      if df_mart.loc[i, val] <= quantile.iloc[7] :
        df_mart.loc[i, 'grp'] = 1
      elif df_mart.loc[i, val] <= quantile.iloc[14] :
        df_mart.loc[i, 'grp'] = 2
      else : df_mart.loc[i, 'grp'] = 3

    # ▶ non target labeling
    df_mart['n_target'] = np.where(df_mart['target']==0, 1, 0)

    # ▶ target , non-target count
    iv = df_mart.groupby('grp')[['target', 'n_target']].sum().reset_index()

    # ▶ good_pct / bad_pct 계산
    iv['good_pct'] = iv['target'] / iv['target'].sum()
    iv['bad_pct'] = iv['n_target'] / iv['n_target'].sum()

    # ▶ t_ratio
    iv['t_ratio'] = iv['target'] / (iv['target'] + iv['n_target'])

    iv['val'] = val

    for i in range(len(iv['grp'])) :
      iv.loc[i, 'iv'] = (iv.loc[i,'good_pct'] - iv.loc[i,'bad_pct']) * math.log(iv.loc[i,'good_pct']/iv.loc[i,'bad_pct'])

    iv_df.append(iv)

iv_df=pd.concat(iv_df)
iv_df
````

> **Feature Engineering이란?**

```
* 모델링의 목적인 성능향상을 위해 Feature를 생성, 선택, 가공하는 일련의 모든 활동
```


> **Target 변수와의 의미있는 변수 선택하는 방법**

```
* 차원이 너무 많은 경우에는 차원을 Domain 지식을 활용하여 축소할 것
```

> **IV (Information Value)**

```
* Feature 하나가 Good(Target)과 Bad(Non-target)을 잘 구분해줄 수 있는지에 대한 정보량을 표현하는 방법론
* IV 수치가 클수록 Target과 Non-target을 잘 구분할 수 있는 정보량이 많은 Feature이고, IV 수치가 작을수록 정보량이 적은 Feature
* (Target data 구성비 - Non-target data 구성비) * WoE
* WoE(Weight of Evidence) = ln(Target data 구성비 / Non-target data 구성비)
* ln(자연로그)를 취하는 이유? : WoE의 최대, 최소값의 범위를 맞춰주기 위해 (1~∞ , -∞~1)
```
