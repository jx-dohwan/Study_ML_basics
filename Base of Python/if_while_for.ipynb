## if Statement

```python
list_type = [1,2,3,4]
dict_type = {'a':1, 'b':2, 'c':3, 'd':4}
str_type = "testing"
```

```python
if type(list_type) == list:
  print("This is a list")
```

```python
if type(dict_type) == dict:
  print("This is a dictionary")
```

```python
if type(str_type) == str:
  print("This is a string")
```

```python
if dict_type['a'] == 1:
  print(f"key:a value:{dict_type['a']}")
```

```python
if type(list_type) == list and type(dict_type) == dict:
  print("back")
```

```python
if 1 and 1:
  print('a')
```

```python
if 1 and 2:
  print('a')
```

```python
if 1 & 2:
  print('a')
```

## while

```python
i = 1
while(1):
  i += 1
  if i == 10:
    print(i)
    break
```

```python
i = 1
while True:
  i += 1
  if i == 10:
    print(i)
    break
```

## try except
